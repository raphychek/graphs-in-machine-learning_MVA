# Question 1.1

The optionnal parameter $\textbf{gen_pam}$ correspond to the distance to which the first data element of the Gaussian blobs generated by $\textbf{skd.make_blobs}$ is shifted on the abscissa axis from the previous far right element.

We can see this in the next image, for $\textbf{gen_pam} \in [-1,-3,-5,-10,1,3,5,10]$ :

```python
fig = plt.figure(figsize=(12,5))
gen_pams = [[-1,-3,-5,-10],[1,3,5,10]]

for i in range(2):
    for j in range(4):
        X, Y = worst_case_blob(100, gen_pams[i][j])
        plt.subplot(2,4,4*i+j+1)
        plt.scatter(X[:,0], X[:,1], c='g', label='gen_pam= ' + str(gen_pams[i][j]))
        plt.scatter(X[-1,0], X[-1,1], c='r')
        plt.legend()
```

# Question 1.2

If $\textbf{gen_pam}$ is big enough, the graph goes from a one connected component graph to a two connected component graph, as the worst blob now is way more far than the other blobs from the center of the Gaussian.

Moreover, as $\epsilon$ is chosen as a function of the maximal value of distance between two linked vertexes whose edge is present in the minimal spanning tree of the fully connected graph, the higher this parameter goes, the more dense is the graph matrix - we can know it by the color of the matrix. It seems to reach a density saturation around when $|\textbf{gen_pam}|>\sigma^2$, which makes sense knowing than when the blob is farther from all of the other blobs from $\sigma^2$, it shouldn't be a neighbor of any blobs anymore.

In term of value, we notice that the closest $\epsilon$ is from 0, the densier is the graph.

We can observe this in the next figures, for $\textbf{gen_pam} \in [0,0.2,0.4,0.6,0.8,1,3]$:

```python
gen_pams = [-5, -3, -1,-0.5,0, 0.5, 1, 3,5]
epsilons = []
i,j=0,0

for gen_pam in gen_pams:
    print("gen_pam =",gen_pam)
    epsilon,X,Y,W = how_to_choose_epsilon(gen_pam = gen_pam)
    epsilons.append(epsilon)
    plot_graph_matrix(X,Y,W)
    i+=1
    j+=1
    

df = pd.DataFrame({'gen_pam':gen_pams, 'epsilon':epsilons})

df
```

# Question 1.3

When the data is clean, so when $\textbf{gen_pam}$ is very low or null, an $\epsilon$-graph is the easiest to compute and is relevent enough. When $\textbf{gen_pam}$ is high, the $\epsilon$ parameter isn't relevant anymore because of our criteria of selection.
Thereby, in this case, constructing a $k-nn$-graph is best, as the results are still relevant. 

We can observe this in the next figures, for $\textbf{gen_pam} \in [0,1,10,100]$ and $k=5$:

```python
for gp in [0,1,10,100]:
    _,X,Y,W=how_to_choose_epsilon(gp,5)
    plot_graph_matrix(X,Y,W)
```
